```python
import pandas as pd

try:
    df = pd.read_csv('sentimentdataset.csv')
    display(df.head())
except FileNotFoundError:
    print("Error: 'sentimentdataset.csv' not found. Please ensure the file exists in the current directory or provide the correct path.")
except Exception as e:
    print(f"An unexpected error occurred: {e}")
```

```python
# Examine Data Shape and Structure
print("Data Shape:", df.shape)
print("\nColumn Names:", df.columns.tolist())

# Check Data Types
print("\nData Types:\n", df.dtypes)

# Identify Missing Values
print("\nMissing Values:\n", df.isnull().sum())
print("\nPercentage of Missing Values:\n", (df.isnull().sum() / len(df)) * 100)

# Analyze Sentiment Distribution
sentiment_counts = df['Sentiment'].value_counts()
print("\nSentiment Distribution:\n", sentiment_counts)

import matplotlib.pyplot as plt
plt.figure(figsize=(8, 6))
sentiment_counts.plot(kind='bar')
plt.title('Distribution of Sentiments')
plt.xlabel('Sentiment')
plt.ylabel('Frequency')
plt.show()


# Investigate Numerical Features
numerical_features = ['Retweets', 'Likes']
print("\nDescriptive Statistics for Numerical Features:\n", df[numerical_features].describe())

# Check for Outliers (using box plots)
plt.figure(figsize=(10, 6))
df[numerical_features].boxplot()
plt.title('Box Plots of Numerical Features')
plt.ylabel('Values')
plt.show()

# Preliminary Text Data Analysis
if 'Text' in df.columns:
    df['Text Length'] = df['Text'].astype(str).apply(len)  # Calculate text lengths
    print("\nText Length Statistics:\n", df['Text Length'].describe())

    plt.figure(figsize=(8, 6))
    plt.hist(df['Text Length'], bins=20)
    plt.title('Distribution of Text Lengths')
    plt.xlabel('Text Length')
    plt.ylabel('Frequency')
    plt.show()
else:
    print("\nNo 'Text' column found in the DataFrame.")

# Summarize Findings (This will be done in the next step as text, not code.)
```

```python
# Convert 'Retweets' and 'Likes' to numeric, coercing errors to NaN
df['Retweets'] = pd.to_numeric(df['Retweets'], errors='coerce')
df['Likes'] = pd.to_numeric(df['Likes'], errors='coerce')

# Convert 'Timestamp' to datetime format
df['Timestamp'] = pd.to_datetime(df['Timestamp'], errors='coerce')

# Remove duplicate rows
df.drop_duplicates(inplace=True)

# Handle outliers in 'Retweets' and 'Likes' using IQR method
for col in ['Retweets', 'Likes']:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    df = df[(df[col] >= lower_bound) & (df[col] <= upper_bound)]

# Reset the index after removing rows
df.reset_index(drop=True, inplace=True)

display(df.head())
```

## Data visualization

## Feature engineering

### Subtask:
Engineer relevant features from the existing dataset to improve model performance for subsequent tasks (like clustering or classification, if those were part of the original plan).


**Reasoning**:
Engineer features as instructed, including TF-IDF, engagement ratio, and sentiment encoding.  Combine these into a single feature matrix.  No dimensionality reduction will be performed in this step.



```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.preprocessing import LabelEncoder

# 1. Text Feature Engineering (TF-IDF)
tfidf_vectorizer = TfidfVectorizer(stop_words='english', max_features=1000) # Limit features for demonstration
tfidf_matrix = tfidf_vectorizer.fit_transform(df['Text'])

# 2. Engagement Ratio Feature
df['EngagementRatio'] = (df['Retweets'] + df['Likes']) / len(df)

# 3. Sentiment Encoding
label_encoder = LabelEncoder()
df['EncodedSentiment'] = label_encoder.fit_transform(df['Sentiment'])

# 4. Combine Features
combined_features = pd.concat([pd.DataFrame(tfidf_matrix.toarray()),
                              df[['EngagementRatio', 'EncodedSentiment']]], axis=1)
display(combined_features.head())
```

## Data clustering

Perform K-means clustering on the combined features.


**Reasoning**:
Perform K-means clustering on the combined features and add the cluster labels to the original dataframe.



```python
from sklearn.cluster import KMeans

# Convert column names to strings
combined_features.columns = combined_features.columns.astype(str)

# Initialize KMeans with a specified number of clusters (e.g., 5)
kmeans = KMeans(n_clusters=5, random_state=0)  # You might want to experiment with different n_clusters

# Fit the KMeans model to the combined features
kmeans.fit(combined_features)

# Add the cluster labels to the original dataframe
df['cluster_label'] = kmeans.labels_

# Display the first few rows of the updated dataframe to verify
display(df.head())
```
